Looking at the box plot, it seems that the six variables have outlier present in the variables and the six variables are as follows:


*   **Price**
*  **Max.price**
*   **Horsepower**
*   **MPG.highway**
*   **MPG.city**
*   **Weight**



These outliers value needs to be teated and there are several ways of treating them:

1.**Drop the outlier value**

2.**Replace the outlier value using the IQR**

Created a user definded function for finding the lower and upper range for a variable so that outlier can be treated

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range=Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lrprice,urprice=remove_outlier(df['Price'])
df['Price']=np.where(df['Price']>urprice,urprice,df['Price'])
df['Price']=np.where(df['Price']<lrprice,lrprice,df['Price'])


lrmax,urmax=remove_outlier(df['Max.Price'])
df['Max.Price']=np.where(df['Max.Price']>urmax,urmax,df['Max.Price'])
df['Max.Price']=np.where(df['Max.Price']<lrmax,lrmax,df['Max.Price'])


lrhorsepower,urhorsepower=remove_outlier(df['Horsepower'])
df['Horsepower']=np.where(df['Horsepower']>urhorsepower,urhorsepower,df['Horsepower'])
df['Horsepower']=np.where(df['Horsepower']<lrhorsepower,lrhorsepower,df['Horsepower'])


lrhighway,urhighway=remove_outlier(df['MPG.highway'])
df['MPG.highway']=np.where(df['MPG.highway']>urhighway,urhighway,df['MPG.highway'])
df['MPG.highway']=np.where(df['MPG.highway']<lrhighway,lrhighway,df['MPG.highway'])


lrcity,urcity=remove_outlier(df['MPG.city'])
df['MPG.city']=np.where(df['MPG.city']>urcity,urcity,df['MPG.city'])
df['MPG.city']=np.where(df['MPG.city']<lrcity,lrcity,df['MPG.city'])


lrweight,urweight=remove_outlier(df['Weight'])
df['Weight']=np.where(df['Weight']>urweight,urweight,df['Weight'])
df['Weight']=np.where(df['Weight']<lrweight,lrweight,df['Weight'])

df.shape
```

# **Check for Missing value**

```python
df.isnull().sum()
```

There are missing values present are  given below:



*   **Rear.seat.room**
*   **Luggage.room**
*   **AirBags**

1.Rear.seat.room and Luggage.room are numerical variables and both are not having outlier so we will use mean to replace them.

2.AirBags is categorical variable so we will mode to replace it.



```python
df[df.isnull().sum()[df.isnull().sum()>0].index].dtypes
```

**Replacing NULL values in numerical columns with mean, because our numerical values dont have outlier.**

```python
mean1=df["Rear.seat.room"].mean()
mean2=df["Luggage.room"].mean()

df["Rear.seat.room"].replace(np.nan,mean1,inplace=True)
df["Luggage.room"].replace(np.nan,mean2,inplace=True)
```

**Replacing NULL values in Categorical columns using Mode**

```python
mode1=df["AirBags"].mode().values[0]

df["AirBags"]=df["AirBags"].replace(np.nan,mode1)
```

# **Normalization and Scaling(numerical variables)**

### **Identifying Numerical Variables**

```python
numerical_cols = df.select_dtypes(include=np.number).columns
print("Numerical columns:", numerical_cols)
```

## **Standardization (StandardScaler)**

```python
from sklearn.preprocessing import StandardScaler

scaler_standard = StandardScaler()
df_scaled_standard = df.copy() # Create a copy to avoid modifying the original DataFrame
df_scaled_standard[numerical_cols] = scaler_standard.fit_transform(df[numerical_cols])

print("DataFrame after Standardization:")
display(df_scaled_standard.head())
```

## **Normalization (MinMaxScaler)**

```python
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
df_scaled_minmax = df.copy() # Create a copy to avoid modifying the original DataFrame
df_scaled_minmax[numerical_cols] = scaler_minmax.fit_transform(df[numerical_cols])

print("DataFrame after Normalization:")
display(df_scaled_minmax.head())
```

**Now you have two new DataFrames, df_scaled_standard and df_scaled_minmax, with the numerical columns scaled using Standardization and Normalization respectively.**
